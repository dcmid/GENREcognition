## Preprocess Features Data

```python
# Because all the array data are acutally stored as string in the csv file, we transform it to true array using literal_eval
from ast import literal_eval

raw_data['zero_crossing (zero_crossing_sum & total_frame)'] = raw_data['zero_crossing (zero_crossing_sum & total_frame)'].apply(literal_eval)
raw_data['spectral_centroids'] = raw_data['spectral_centroids'].apply(literal_eval)
raw_data['spectral_rolloff'] = raw_data['spectral_rolloff'].apply(literal_eval)
```

```python
# mfccs data turn out to be too large to be transformed

# raw_data['mfccs'] = raw_data['mfccs'].apply(literal_eval)
```

```python
# Preprocess features data for training.  
"""
Here we calculate:
1. zero_crossing_rate
2. spectral_centroids_mean
3. spectral_centroids_std
4. spectral_rolloff_mean
5. spectral_rolloff_std
"""

# // TODO: Feel free to add more features that can be used to train data


raw_data['zero_crossing_rate'] = raw_data['zero_crossing (zero_crossing_sum & total_frame)'].apply(lambda x: x[0]/x[1])
raw_data['spectral_centroids_mean'] = raw_data['spectral_centroids'].apply(np.mean)
raw_data['spectral_centroids_std'] = raw_data['spectral_centroids'].apply(np.std)
raw_data['spectral_rolloff_mean'] = raw_data['spectral_rolloff'].apply(np.mean)
raw_data['spectral_rolloff_std'] = raw_data['spectral_rolloff'].apply(np.std)


```

## Preprocess for Training

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(raw_data.genre_top)
raw_data['categorical_genre_top'] = le.transform(raw_data.genre_top)
```

## Train and Predict using SVM

```python
# We use SVM to train and predict genre based on different features combination and try to find out which combination works better

from sklearn import svm
from sklearn.metrics import accuracy_score

def cal_accuracy_by_selected_features(features):
    X = raw_data[features]
    y = raw_data['categorical_genre_top']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    clf = svm.SVC()
    clf.fit(X_train, y_train)
    clf_predict = clf.predict(X_test)
    my_accuracy = accuracy_score(y_test, clf_predict, normalize=False) / float(y_test.size)
#     print('Features: ', features)
#     print('Accuracy: ', my_accuracy)
#     print('-------------')
    return my_accuracy
```

```python
import itertools

all_features = ['zero_crossing_rate','spectral_centroids_mean','spectral_centroids_std','spectral_rolloff_mean','spectral_rolloff_std']

all_comb = []

for length in range(1, len(all_features)):
    for i in itertools.combinations(all_features, length):
        all_comb += [list(i)]
```

```python
result = []
for features in all_comb:
    accu = cal_accuracy_by_selected_features(features)
    result += [[features, accu]]
```

```python
result.sort(key=lambda x: -x[1])
for r in result:
    print(r[0])
    print(r[1])
```

## Train and Predict using RandomForest

```python
# We use RandomForest to train and predict genre based on different features combination and try to find out which combination works better


from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

def cal_accuracy_by_selected_features(features):
    X = raw_data[features]
    y = raw_data['categorical_genre_top']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    clf = RandomForestClassifier(max_depth=2, random_state=0)
    clf.fit(X_train, y_train)
    clf_predict = clf.predict(X_test)
    my_accuracy = accuracy_score(y_test, clf_predict, normalize=False) / float(y_test.size)

    return my_accuracy
```

```python
result = []
for features in all_comb:
    accu = cal_accuracy_by_selected_features(features)
    result += [[features, accu]]
```

```python
result.sort(key=lambda x: -x[1])
for r in result:
    print(r[0])
    print(r[1])
```

## Train and Predict using Keras

```python
# We use Keras to train and predict genre based on different features combination and try to find out which combination works better

from keras import models
from keras import layers
def cal_accuracy_by_selected_features(features):
    X = raw_data[features]
    y = raw_data['categorical_genre_top']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    model = models.Sequential()
    model.add(layers.Dense(256, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(10, activation='softmax'))
    model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
    history = model.fit(X_train, y_train,
                    epochs=20,
                    batch_size=128)
    test_loss, test_acc = model.evaluate(X_test,y_test)
 
    return test_acc
```

```python
result = []
for features in all_comb:
    accu = cal_accuracy_by_selected_features(features)
    result += [[features, accu]]
```
